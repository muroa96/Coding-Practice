```python
# 신고 결과 받기
def solution(id_list, report, k):
    answer = []
    #1. 중복 신고 제거
    report = list(set(report))
    #2. 신고유저와 신고받은 유저로 report 리스트 나누기(공백기준)
    ls  = [i.split(' ') for i in report]
    #3. 정지당한 유저 찾기
    reported_user =[i[1] for i in ls]
    banned_user = [i for i in id_list if reported_user.count(i)>=k]
    #4. 메일 보내기
    ls = [i for i in ls if i[1] in banned_user]
    email = [i[0] for i in ls]
    answer = [email.count(i) for i in id_list]
    return answer
```

```python
#[1차] 다트게임
def solution(dartResult):
    answer = 0
    #1. 점수(숫자) 인덱스 찾기
    score_list = [str(i) for i in range(11)]
    bonus_list = ['S', 'D', 'T']
    option_list = ['*', '#']
    index_list = [i for i in range(len(dartResult)) if dartResult[i] in score_list]
    index_list = [i for i in index_list if i-1 not in index_list] # 점수 10점이 있는 경우
    #2. 점수들 나누기
    first_score = dartResult[0:index_list[1]]
    second_score = dartResult[index_list[1]:index_list[2]]
    third_score = dartResult[index_list[2]:]
    
    fs = 0
    ss = 0
    ts = 0
    if '10' in first_score:
        fs+=10
    else:
        fs+=int(first_score[0])
    if '10' in second_score:
        ss+=10
    else:
        ss+=int(second_score[0])
    if '10' in third_score:
        ts+=10
    else:
        ts+=int(third_score[0])
    #3. 점수 계산하기
    ## first score
    for i in range(len(first_score)):
        if first_score[i] in bonus_list: #S, D, T
            fs = fs**(bonus_list.index(first_score[i])+1)
        if first_score[i] == '#':
            fs = -fs
    if '*' in first_score:
        fs = fs*2
    ##second score
    for j in range(len(second_score)):
        if second_score[j] in bonus_list: #S, D, T
            ss = ss**(bonus_list.index(second_score[j])+1)
        if second_score[j] == '#':
            ss = -ss
        if second_score[j] == '*':
            fs = fs*2
            ss = ss*2
    ##third score
    for k in range(len(third_score)):
        if third_score[k] in bonus_list: #S, D, T
            ts = ts**(bonus_list.index(third_score[k])+1)
        if third_score[k] == '#':
            ts = -ts
        if third_score[k] == '*':
            ss = ss*2
            ts = ts*2
    answer = fs+ss+ts
    return answer
```

```python
#큰 수 만들기
# 조합을 활용한 풀이는 timeout이 된다.
from itertools import combinations
def solution(number, k):
    answer = ''
    num_list = [int(i) for i in number]
    all_nums = [i for i in combinations(num_list, len(number)-k)]
    for i in max(all_nums):
        answer += str(i)
    return answer


```
